# Question 1- Generating the car matrix 
Under the function named generate_car_matrix write a logic that takes the dataset-1.csv as a DataFrame. Return a new DataFrame that follows the following rules:

values from id_2 as columns

values from id_1 as index

dataframe should have values from car column

diagonal values should be 0.

Sample result dataframe:

```python
def generate_car_matrix(dataset):
    #pivot the df to create a matrix with 'id_1' as index, 'id_2' as columns, and 'car' as values
    car_matrix = dataset.pivot(index='id_1', columns='id_2', values='car').fillna(0)

    #seta diagonal values to 0
    for i in car_matrix.index:
        car_matrix.at[i, i] = 0

    return car_matrix

car_matrix_result = generate_car_matrix(dataset)
```

# Question2 -Car Type Count Collection

***Create a Python function named get_type_count that takes the dataset-1.csv as a DataFrame. Add a new categorical column car_type based on values of the column car:***

low for values less than or equal to 15,

medium for values greater than 15 and less than or equal to 25,

high for values greater than 25.

Calculate the count of occurrences for each car_type category and return the result as a dictionary. Sort the dictionary 
alphabetically based on keys.

```python
def get_type_count(dataset):
    conditions = [
        (dataset['car'] <= 15),
        (dataset['car'] > 15) & (dataset['car'] <= 25),
        (dataset['car'] > 25)
    ]
    choices = ['low', 'medium', 'high']
    dataset['car_type'] = pd.cut(dataset['car'], bins=[-float('inf'), 15, 25, float('inf')], labels=choices)

    #calculate the count of occurrences for each 'car_type' category
    type_counts = dataset['car_type'].value_counts().to_dict()
    sorted_type_counts = dict(sorted(type_counts.items()))

    return sorted_type_counts


type_count_result = get_type_count(dataset)
```

# Question 3: Bus Count Index Retrieval
Create a Python function named get_bus_indexes that takes the dataset-1.csv as a DataFrame. 
The function should identify and return the indices as a list (sorted in ascending order) where the bus values are greater than twice the mean value of the bus column in the DataFrame.

```python


def get_bus_indexes(dataset):
    #calculate the mean v
    mean_bus = dataset['bus'].mean()
    print("mean value fro bus", mean_bus)

    #her we are checing 'bus' values are greater than twice the mean
    bus_indexes = dataset[dataset['bus'] > 2 * mean_bus].index.tolist()
    bus_indexes.sort()
    return bus_indexes
bus_indexes_result = get_bus_indexes(dataset)
```

# Question 4: Route Filtering
Create a python function filter_routes that takes the dataset-1.csv as a DataFrame. The function should return the sorted list of values of column route for which the average of values of truck column is greater than 7.

```python
def filter_routes(dataset):
    #group by route and calculate the mean of the truck column
    route_avg_truck = dataset.groupby('route')['truck'].mean()

    # here ii am creating routes where the average of truck column is greater than 7
    filtered_routes = route_avg_truck[route_avg_truck > 7].index.tolist()
    filtered_routes.sort()
    return filtered_routes

filtered_routes_result = filter_routes(dataset)
```

```python
print(filtered_routes_result)
```

If a value in the DataFrame is greater than 20, multiply those values by 0.75,
